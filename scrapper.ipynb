```python
service = Service(executable_path='./chromedriver-linux64/chromedriver')
driver = webdriver.Chrome(service=service)
driver.get('https://www.urt.gov.co/sentencias-por-departamento/-/document_library/SintN68izetz/view/127568?_com_liferay_document_library_web_portlet_DLPortlet_INSTANCE_SintN68izetz_navigation=home&_com_liferay_document_library_web_portlet_DLPortlet_INSTANCE_SintN68izetz_curFolder=&_com_liferay_document_library_web_portlet_DLPortlet_INSTANCE_SintN68izetz_deltaFolder=&_com_liferay_document_library_web_portlet_DLPortlet_INSTANCE_SintN68izetz_deltaEntry=20&_com_liferay_document_library_web_portlet_DLPortlet_INSTANCE_SintN68izetz_orderByCol=modifiedDate&_com_liferay_document_library_web_portlet_DLPortlet_INSTANCE_SintN68izetz_orderByType=desc&p_r_p_resetCur=false&_com_liferay_document_library_web_portlet_DLPortlet_INSTANCE_SintN68izetz_curEntry=1')
time.sleep(6)
```

```python
initial_links = driver.find_elements(By.CSS_SELECTOR, "td.table-cell-expand a")
```

```python
links_dict = {}
```

```python
link_hrefs = [element.get_attribute('href') for element in initial_links]
link_texts = [element.text for element in initial_links]
```

```python
for text, href in zip(link_texts, link_hrefs):
    driver.get(href)

    new_links = driver.find_elements(By.CSS_SELECTOR, "td.table-cell-expand a")
    new_hrefs = [link.get_attribute('href') for link in new_links]
    new_link_texts = [element.text for element in new_links]

    if text not in links_dict:
        links_dict[text] = {}

    for new_text, new_href in zip(new_link_texts, new_hrefs):
        if new_text not in links_dict[text]:
            links_dict[text][new_text] = [] 
        links_dict[text][new_text].append(new_href) 
```

```python
links_dict
```

```python
documents_dict = {}
```

```python
for key, value in links_dict.items():

    if key not in documents_dict:
        documents_dict[key] = {}

    for sub_key, sub_value in value.items():

        if sub_key not in documents_dict[key]:
            documents_dict[key][sub_key] = []

        for href in sub_value:
            driver.get(href)
            time.sleep(2)

            anchors = driver.find_elements(By.CSS_SELECTOR, "td.table-cell-expand.table-cell-minw-200.table-title.lfr-title-column a")
            links = [anchor.get_attribute('href') for anchor in anchors]

            documents_dict[key][sub_key].extend(links)     
            
```

```python
documents_dict
```

```python
filename = 'links.txt'

# Write to file in a readable format
with open(filename, 'w') as file:
    json.dump(documents_dict, file, indent=4, ensure_ascii=False)
```
